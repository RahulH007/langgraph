```python
from langgraph.graph import StateGraph , START, END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv

```

```python
load_dotenv()
```

```python
model=ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=1.0,  # Gemini 3.0+ defaults to 1.0
    max_tokens=None
)
```

```python
class BlogState(TypedDict):
    title : str
    outline : str
    content : str
```

```python
def create_outline(state : BlogState) -> BlogState:
    
    title = state['title']
    prompt =f"Generate aa detailed outline for a blog on the topic : {title}"
    response = model.invoke(prompt).content
    state['outline'] = response

    return state
```

```python
def create_blog(state: BlogState) -> BlogState:
    outline = state['outline']
    prompt = f"Generate a blog post based on the following outline : {outline}"
    response = model.invoke(prompt).content
    state['content'] = response
    return state
```

```python
graph = StateGraph(BlogState)

graph.add_node('create_outline',create_outline)
graph.add_node('create_blog',create_blog)

graph.add_edge(START,'create_outline')
graph.add_edge('create_outline','create_blog')
graph.add_edge('create_blog',END)

app = graph.compile()
```

```python
input = { 'title' : "The future of AI in Colleges"}

output = app.invoke(input)

print(output['outline'])
```
